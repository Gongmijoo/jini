*<이상치를 측정해낼 수 있는 공식>*

q1(1분위=0.25)와 q3(3분위=0.75)의 boxplot 범위를 만들어주고

그 범위를 가지고 q1과q3의 형상을 만든다.

그리고 그 안에 데이터를 넣어준다.


```python
q1 = df_salary["salary"].quantile(q=0.25)
q3 = df_salary["salary"].quantile(q=0.75)
IQR = q3-q1
IQR
#boxplot의 범위 만들기 작업/0.25는 1분위,0.75는 3분위 수/항상 3분위-1분위를 하면 boxplot의 범위가 나옴
```

```python
data_IQR = df_salary[(df_salary["salary"] <q3 + IQR * 1.5) & (df_salary["salary"]> q1 -IQR * 1.5)]
data_IQR
#boxplot의 범위를 가지고 q3와 q1의 형상이 만들어짐, 그리고 그 안에 데이터를 넣음
#q3+(q3-q1)*1.5/q1-(q3-q1)*1.5
```

```python
data_IQR["salary"].hist() #salary_boxplot의 히스토리그램/정규분포와 비슷하게 그려짐
```

```python
data_IQR = df_salary[(df_salary["salary"] > q3 + IQR * 1.5)]
data_IQR
#q3+IQR*1.5 보다 이상인 이상치값에 관련해서 나온 값 리스트/salary쪽이 변수,25~쪽이 인덱스
```

t-test

"평균비교에 대한 가설을 적어라"

H0:A,B 두집단은 같다(귀무가설)=

H1:A,B 두집단은 다르다(대립가설)=/

```python
from scipy import stats
```

```python
#일표본 t-test
print(df2.총_매출액.mean())

print(stats.ttest_1samp(df2["총_매출액"],7700000))
```

```python
#독립표본 t-test

no_claim=df2[df2["클레임접수여부"]==0]
no_claim
claim=df2[df2["클레임접수여부"]==1]
claim
```

```python
no_claim.방문빈도.mean()
```

```python
claim.방문빈도.mean()
```

```python
print(stats.ttest_ind(no_claim["방문빈도"],claim["방문빈도"],equal_var=False)) #equal_var=False : 등분산 가정없이도 ttest가능
```
